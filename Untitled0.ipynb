## Ejercicio 1: Sistema de riego inteligente
##### Una finca agrícola desea automatizar el proceso de riego utilizando datos de humedad del suelo y pronóstico de lluvia. Se pide desarrollar un programa que analice los niveles de humedad diaria de una semana (expresados en porcentaje) y determine si es necesario activar el sistema de riego cada día. Para ello, el sistema debe considerar un umbral de humedad óptima y un margen de seguridad definidos como constantes, y evitar regar si se pronostica lluvia. El programa deberá recorrer los siete días de la semana con un bucle, aplicar condicionales para decidir cuándo regar y mostrar un resumen final con el número total de riegos realizados y el ahorro de agua estimado.

```python
"""sistema de riego inteligente"""

dias = ["lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"]

NIVEL_HUMENDAD = 18
total_riegos = 0
ahorro_agua = 0

for dia in dias:
    humedad = int(input(f"Ingrese el nivel de humedad (%) para {dia}: "))
    pronostico = input(f"¿Se pronostica lluvia para {dia}? (si/no): ").lower()

    if humedad < NIVEL_HUMENDAD and pronostico == "no":
        print(f"Se activa el riego el {dia} (humedad baja y sin lluvia).")
        total_riegos += 1
        ahorro_agua -= 10
    elif humedad < NIVEL_HUMENDAD and pronostico == "si":
        print(f"No se riega el {dia} porque hay pronóstico de lluvia.")
        total_riegos += 1
        ahorro_agua += 10
    elif humedad >= NIVEL_HUMENDAD:
        print(f"No es necesario regar el {dia}, humedad suficiente.")
        total_riegos += 1
        ahorro_agua += 10
    else:
        print(f"Datos no válidos para el {dia}.")

print("Resumen semanal:")
print(f"Total de riegos realizados: {total_riegos}")
print(f"Ahorro estimado de agua: {ahorro_agua} litros")


```

## Ejercicio 2: Clasificación de calidad de cosecha
##### Una cooperativa frutícola necesita clasificar sus lotes de frutas según la calidad, en base al peso promedio y al porcentaje de piezas defectuosas. Se requiere crear un programa que, para varios lotes ingresados por el usuario, calcule el peso promedio de una muestra y determine el porcentaje de defectos. En función de estos valores, el sistema debe asignar una categoría: “A” para productos de alta calidad, “B” para calidad media y “Rechazo” para los que no cumplen los estándares. El programa debe usar funciones para los cálculos, bucles para procesar varios lotes y condicionales para establecer la categoría final, mostrando un reporte con los resultados.

```python
def calcular_peso_promedio(pesos):
    return sum(pesos) / len(pesos)

def calcular_porcentaje_defectos(defectuosas, total):
    return (defectuosas / total) * 100

def clasificar_lote(peso_promedio, porcentaje_defectos):
    if peso_promedio >= 20 and porcentaje_defectos <= 20:
        return "A"
    elif 15 <= peso_promedio < 20 and porcentaje_defectos <= 40:
        return "B"
    else:
        return "Rechazo"


resultados = []


num_lotes = int(input("¿Cuántos lotes desea ingresar? "))


for lote in range(1, num_lotes + 1):
    print(f"\n--- Lote {lote} ---")
    total_piezas = int(input("Ingrese la cantidad total de piezas en la muestra: "))
    piezas_defectuosas = int(input("Ingrese la cantidad de piezas defectuosas: "))

    pesos = []
    for i in range(total_piezas):
        peso = float(input(f"Ingrese el peso de la pieza {i + 1}: "))
        pesos.append(peso)

    peso_promedio = calcular_peso_promedio(pesos)
    porcentaje_defectos = calcular_porcentaje_defectos(piezas_defectuosas, total_piezas)
    categoria = clasificar_lote(peso_promedio, porcentaje_defectos)

    resultados.append({
        "lote": lote,
        "peso_promedio": peso_promedio,
        "porcentaje_defectos": porcentaje_defectos,
        "categoria": categoria
    })

45454
print("\n=== REPORTE FINAL ===")
for resultado in resultados:
    print(f"Lote {resultado['lote']}:")
    print(f"  Peso promedio: {resultado['peso_promedio']:.2f}")
    print(f"  % de defectos: {resultado['porcentaje_defectos']:.2f}%")
    print(f"  Categoría: {resultado['categoria']}")
    print("-" * 30)
```

##Ejercicio 3: Planificación de rotación de cultivos
##### Un ingeniero agrónomo busca diseñar una herramienta que ayude a planificar la rotación de cultivos durante el año y calcular el consumo total de nitrógeno (N) en el suelo. El programa deberá recibir una lista con los cultivos sembrados cada mes (por ejemplo, maíz, soya, trigo o descanso) y usar constantes que indiquen el consumo mensual de nitrógeno por cultivo. Mediante funciones, el sistema debe sumar el consumo total de N y verificar si existen más de tres meses consecutivos del mismo cultivo, recomendando alternar con un cultivo diferente o un período de descanso.

```python
import random
Cultivos = ["zanahorias","papas","mazorcas","maiz","descanzo","descanzo","uvas","fresas","cocos","cebollas","tomate","naranjas"]
cultivo_N = {
    "zanahorias": 27,
    "papas": 15,
    "mazorcas": 45,
    "maiz": 21,
    "descanzo": 0,
    "uvas": 32,
    "fresas": 61,
    "cocos": 11,
    "cebollas": 13,
    "tomate": 31,
    "naranjas": 98,
}
random.shuffle(Cultivos)




def sumar_nitrogeno(Cultivos, consumo_por_cultivo):
    total = 0
    for cultivo in Cultivos:
        total += consumo_por_cultivo.get(cultivo, 0)
    return total

def verificar_rotacion(cultivos):
    anterior = None
    contador = 0
    advertencias = []

    for mes in cultivos:
        if mes == anterior:
            contador += 1
            if contador >= 3:
                advertencias.append(mes)
        else:
            anterior = mes
            contador = 1

    return list(set(advertencias))

print(" Planificación anual de cultivos:")
for i, cultivo in enumerate(Cultivos, start=1):
    print(f"Mes {i}: {cultivo}")


total_n = sumar_nitrogeno(Cultivos, cultivo_N)
print(f" Consumo total de nitrógeno del año: {total_n} unidades")


advertencias = verificar_rotacion(Cultivos)
if advertencias:
    print(" Advertencia de rotación inadecuada:")
    for cultivo in advertencias:
        print(f" - Más de 3 meses consecutivos de {cultivo}. Se recomienda alternar o incluir descanso.")
else:
    print(" Buena rotación de cultivos. No se detectaron problemas.")





```

## Ejercicio 4: Control de inventario de un software agrícola
##### Una empresa que desarrolla software para el agro necesita un módulo que controle el inventario de insumos (fertilizantes, semillas, pesticidas). El programa debe permitir registrar la cantidad inicial de cada producto y simular la salida de unidades según las ventas semanales. Con el uso de bucles y condicionales, el sistema debe actualizar los valores y emitir una alerta cuando algún producto quede por debajo de un nivel mínimo definido como constante. Además, se debe incluir una función que calcule el valor total del inventario restante y muestre un reporte con el stock actualizado.



```python

STOCK_MINIMO = 5

inventario = {
    "fertilizantes": [50, 10.5],
    "semillas": [100, 2.75],
    "pesticidas": [30, 20.0]
}

def actualizar_inventario(inventario):
    print("\n-- Simulación de ventas semanales --")
    for producto in inventario:
        cantidad_actual = inventario[producto][0]
        while True:
            try:
                venta = int(input(f"Ingrese unidades vendidas de {producto} (stock actual {cantidad_actual}): "))
                if 0 <= venta <= cantidad_actual:
                    inventario[producto][0] -= venta
                    break
                else:
                    print(f"Error: La venta debe ser entre 0 y {cantidad_actual}.")
            except ValueError:
                print("Por favor, ingrese un número válido.")

def calcular_valor_total(inventario):
    total = 0
    for producto, (cantidad, precio) in inventario.items():
        total += cantidad * precio
    return total

def mostrar_reporte(inventario):
    print("\n--- Reporte actualizado de inventario ---")
    for producto, (cantidad, precio) in inventario.items():
        print(f"{producto.capitalize()}: {cantidad} unidades disponibles (Precio unitario: ${precio})")
        if cantidad < STOCK_MINIMO:
            print(f"  ¡Alerta! Stock bajo para {producto}. Quedan solo {cantidad} unidades.")
    total_valor = calcular_valor_total(inventario)
    print(f"\nValor total del inventario restante: ${total_valor:.2f}")

def main():
    print("=== Control de Inventario de Insumos Agrícolas ===")
    actualizar_inventario(inventario)
    mostrar_reporte(inventario)

if __name__ == "__main__":
    main()
```

##Ejercicio 5: Evaluación de desempeño de maquinaria
##### Una cooperativa agrícola quiere medir el rendimiento de sus tractores en campo. Crea un programa que permita ingresar los datos de varios tractores: horas trabajadas, litros de combustible consumidos y superficie cubierta. El programa debe calcular el rendimiento por hora y el consumo de combustible por hectárea. Mediante condicionales, debe clasificar el desempeño como “Excelente”, “Aceptable” o “Deficiente”, según los valores obtenidos. Usa funciones para los cálculos, bucles para procesar varios tractores y constantes para los límites de clasificación. Finalmente, muestra un informe con los resultados de cada máquina.


```python

horas_trabajadas = 0
litros_combustible = 0
superficie_cubierta = 0


RENDIMIENTO_EXCELENTE = 10
RENDIMIENTO_ACEPTABLE = 7


def calcular_rendimiento_hora(horas_trabajadas, superficie_cubierta):
    if horas_trabajadas == 0:
        return 0
    return superficie_cubierta / horas_trabajadas


def calcular_consumo_hectarea(litros_combustible, superficie_cubierta):
    if superficie_cubierta == 0:
        return 0
    return litros_combustible / superficie_cubierta


def clasificar_rendimiento(rendimiento):
    if rendimiento >= RENDIMIENTO_EXCELENTE:
        return "Excelente"
    elif rendimiento >= RENDIMIENTO_ACEPTABLE:
        return "Aceptable"
    else:
        return "Deficiente"

horas_trabajadas = float(input("Horas trabajadas: "))
litros_combustible = float(input("Litros de combustible consumidos: "))
superficie_cubierta = float(input("Superficie cubierta (en hectáreas): "))


rendimiento = calcular_rendimiento_hora(horas_trabajadas, superficie_cubierta)
consumo = calcular_consumo_hectarea(litros_combustible, superficie_cubierta)
clasificacion = clasificar_rendimiento(rendimiento)


print("\n--- Informe del Tractor ---")
print(f"Rendimiento por hora: {rendimiento:.2f} ha/h")
print(f"Consumo por hectárea: {consumo:.2f} L/ha")
print(f"Clasificación: {clasificacion}")
```
